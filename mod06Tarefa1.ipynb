```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

%matplotlib inline
%matplotlib notebook

data = pd.read_csv("dados_nascimento.csv")
data.head
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df = pd.DataFrame(data[["APGAR1","APGAR5","SEXO","GESTACAO","ESTCIVMAE"]])



plt.close('all')
plt.rc("figure", figsize = (25,8))
fig , axes = plt.subplots(2,3)

sns.barplot(ax = axes[0,0],x = "APGAR1", y="SEXO" ,data = df)
sns.barplot(ax = axes[0,1],x = "APGAR1", y="GESTACAO" ,data = df)
sns.barplot(ax = axes[0,2],x = "APGAR1", y="ESTCIVMAE" ,data = df)
sns.barplot(ax = axes[1,0],x = "APGAR5", y="SEXO" ,data = df)
sns.barplot(ax = axes[1,1],x = "APGAR5", y="GESTACAO" ,data = df)
sns.barplot(ax = axes[1,2],x = "APGAR5", y="ESTCIVMAE" ,data = df)
```
